# **Train Test Split**

```python
from tensorflow.keras.utils import to_categorical

Y = to_categorical(Y)
Y
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=500)
```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

# **Creating a New Model Using Transfer Learning**

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(9000,input_shape=[15],activation='tanh'))

model.add(Dense(3,activation='softmax'))

model.compile(optimizer='adam',loss='categorical_crossentropy',metrics=['accuracy'])
```

```python
from tensorflow.keras.callbacks import EarlyStopping

earlystop = EarlyStopping(monitor='val_loss',patience=5)
```

```python
model.fit(X_train,Y_train,validation_data=(X_test,Y_test),epochs=100,callbacks=[earlystop])
```

```python
loss_df = pd.DataFrame(model.history.history)
loss_df.plot()
```

```python
Y_pred = np.round(model.predict(X_test))
```

```python
from sklearn.metrics import classification_report

print(classification_report(Y_test,Y_pred))
```
